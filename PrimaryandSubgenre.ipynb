```python
csv_file = 'data_w_genres.csv'

spotify = pd.read_csv(csv_file, header=0, encoding='utf-8')
```

```python
spotify.head()
```

```python
# changing the metadata in genre cells since there are many sub genres. making it simplier. 

# jazz primary and secondary genre
spotify.loc[spotify['genres'].str.contains('jazz'), 'primary_genres'] = 'jazz'
spotify.loc[spotify['genres'].str.contains('cool jazz'), 'secondary_genres'] = 'cool jazz'

# spotify.loc[spotify['genres'].str.contains('alternative'), 'new_genres'] = 'alternative'
# spotify.loc[spotify['genres'].str.contains('r&b'), 'new_genres'] = 'r&b'


```

```python
# classical primary and secondary genre
spotify.loc[spotify['genres'].str.contains('classical'), 'primary_genres'] = 'classical'
spotify.loc[spotify['genres'].str.contains('performance'), 'secondary_genres'] = 'performance'
```

```python
# salsa primary and secondary genre
spotify.loc[spotify['genres'].str.contains('latin'), 'primary_genres'] = 'latin'
spotify.loc[spotify['genres'].str.contains('salsa'), 'secondary_genres'] = 'salsa'

```

```python
spotify.loc[spotify['genres'].str.contains('metal'), 'primary_genres'] = 'metal'
spotify.loc[spotify['genres'].str.contains('alternative metal'), 'secondary_genres'] = 'alternative metal'
```

```python
spotify.loc[spotify['genres'].str.contains('country'), 'primary_genres'] = 'country'
spotify.loc[spotify['genres'].str.contains('texas country'), 'secondary_genres'] = 'texas country'
```

```python
spotify.loc[spotify['genres'].str.contains('edm'), 'primary_genres'] = 'edm'
spotify.loc[spotify['genres'].str.contains('europop'), 'secondary_genres'] = 'europop'
```

```python
spotify.loc[spotify['genres'].str.contains('christian'), 'primary_genres'] = 'christian'
spotify.loc[spotify['genres'].str.contains('worship'), 'secondary_genres'] = 'worship'
```

```python
spotify.loc[spotify['genres'].str.contains('rap'), 'primary_genres'] = 'rap'
spotify.loc[spotify['genres'].str.contains('trap'), 'secondary_genres'] = 'trap'
```

```python
# new dataset
new_spotify = spotify
new_spotify
```

```python
# need to drop all new_genres cells that are empty and work on creating an even datasheet. 50 titles for all 10 genres. 
# randomizing here
# new_spotify.sample(frac=1)
```

```python
# dropping all rows in new_genre that are empty
compressed = new_spotify.dropna(subset=['secondary_genres'])
compressed
```

```python
compressed['PrimarynSecondary_genres'] = compressed['primary_genres'].str.cat(compressed['secondary_genres'], sep ="/") 
```

```python
# in case there are more fields with NaN
secondcompressed = compressed.dropna(subset=['PrimarynSecondary_genres'])
secondcompressed
```
